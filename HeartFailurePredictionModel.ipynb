### Question 2c
Plot sketch the one dimensional profiles H(ω1, 0), H(ω1, π), and H(0, ω2), H(π, ω2). Use extra points as required. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of omega values
omega = np.linspace(-2*np.pi, 2*np.pi, 400)

# Define the functions
H_w1_0 = 7 + 2 * np.cos(omega)
H_w1_pi = 5 * np.ones_like(omega)
H_0_w2 = 7 + 2 * np.cos(omega)
H_pi_w2 = 5 * np.ones_like(omega)

# Plot H(ω1, 0) and H(0, ω2)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(omega, H_w1_0, label="H(ω1, 0)")
plt.plot(omega, H_w1_pi, linestyle='--', label="H(ω1, π)")
plt.title("H(ω1, 0) and H(ω1, π)")
plt.xlabel("ω1")
plt.ylabel("Magnitude")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(omega, H_0_w2, label="H(0, ω2)")
plt.plot(omega, H_pi_w2, linestyle='--', label="H(π, ω2)")
plt.title("H(0, ω2) and H(π, ω2)")
plt.xlabel("ω2")
plt.ylabel("Magnitude")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
